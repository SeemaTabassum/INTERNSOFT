```python
engine = pyttsx3.init("sapi5")
voices = engine.getProperty('voices')
engine.setProperty("voice",'voices[0].id')
```

```python
def speak(text):
    engine.say(text)
    engine.runAndWait()
```

```python
def wishMe():
    hour=datetime.datetime.now().hour
    if hour >= 0 and hour <= 12:
        speak("hello, Good morning Ms.Seema")
        print("hello, Good morning Ms.Seema")
    elif hour >= 12 and hour <= 18:
        speak("hello, Good afternoon Ms.Seema")
        print("hello, Good afternoon Ms.Seema")
    else:
        speak("it's already night, better go to sleep")
        print("it's already night, better go to sleep")
    
```

```python
def takeCommand():
    r = sr.Recognizer()
    with sr.Microphone() as source:
        print("I AM LISTENING")
        audio = r.listen(source)
        
        try:
            statement = r.recognize_google(audio,language="en-in")
            print(f"user said:{statement}\n")
            
        except Exception as e:
            speak("Pardon me, please say that again")
            return "None"
        return statement
```
